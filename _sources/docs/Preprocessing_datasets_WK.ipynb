# Dataset verwerken

```python
import pandas as pd
```

```python
df_2022 = pd.read_csv('Fifa_world_cup_matches.csv')
df_2022.head()
```

```python
selected_columns = df_2022.loc[:, ['yellow cards team1', 'yellow cards team2', 'red cards team1', 'red cards team2', 'free kicks team1',
       'free kicks team2']]
print(selected_columns)
```

```python
df_2018 = pd.read_csv('FIFA 2018 Statistics.csv')
print(df_2018)
```

```python
# Samenvoegen van rijen in de kolom 'Fouls Committed'
df = df_2018.groupby(df_2018.index // 2).agg({
    'Fouls Committed': 'sum'
}).reset_index(drop=True)

# Resultaat weergeven
display(df)
statistieken = df['Fouls Committed'].describe()
print(statistieken)
```

```python
# kolommen een nieuwe naam geven om ze vervolgens de datasets samen te voegen.
df_2018 = df_2018.rename(columns = {'Team': 'team1'})
df_2018 = df_2018.rename(columns = {'Opponent': 'team2'})
```

```python
namen_kolommen2022 = df_2022.columns
print(namen_kolommen2022)

namen_kolommen2018 = df_2018.columns
print(namen_kolommen2018)
```

```python
merged_df = df_2022.merge(df_2018, on = ['team1', 'team2'], how = 'outer')
display(merged_df)
```

```python
namen_kolommen_merged = merged_df.columns
display(namen_kolommen_merged)
```

```python
merged_df.to_csv('wk_merged_datasets.csv', index=False)
```
